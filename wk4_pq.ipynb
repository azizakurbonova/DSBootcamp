```python
from IPython.display import HTML

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
%matplotlib inline

# Increase default figure and font sizes for easier viewing.
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['font.size'] = 14
```

```python
#Create DF
housing_df = pd.read_csv("boston_housing_data.csv")
dummy_df = pd.DataFrame(np.random.randn(10, 4), 
                  columns=['col1', 'col2', 'col3', 'col4'],
                  index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])
```

```python
#Question 1
housing_df[['ZN', 'INDUS']].plot(figsize=(12,8), style={'ZN': '-g', 'INDUS': '--b'})
plt.style.use("seaborn-v0_8-dark")
```

```python
#Question 2

ax1 = dummy_df[['col1','col2']].plot(kind='bar', figsize=(15,3))

# Set the title.
ax1.set_title('LARGE TITLE', fontsize=50,y=1)

# Move the legend.
ax1.legend(loc=3)

# x-axis labels
ax1.set_ylabel('y-axis info', fontsize=16)

# y-axis labels
ax1.set_xlabel('x-axis info', fontsize=16)

ax2 = dummy_df[['col1','col2']].plot(kind='barh', figsize=(15,3))

# Set the title.
ax2.set_title('LARGE TITLE', fontsize=50,y=1)

# Move the legend.
ax2.legend(loc=1)

# x-axis labels
ax2.set_ylabel('y-axis info', fontsize=16)

# y-axis labels
ax2.set_xlabel('x-axis info', fontsize=16)
```
